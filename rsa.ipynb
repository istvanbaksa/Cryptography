## This Program calculates the private and public exponents for an RSA encryption. It works relatively fast for five- or six-digit-primes.

```python
import math
from numba import njit
import time


p1 = int(input('Please enter first prime number: '))
p2 = int(input('Please enter second prime number: '))

n = p1 * p2
t = (p1-1) * (p2-1)

list1 = []

for i in range(t):
  num1 = math.gcd(i,t)
  num2 = math.gcd(i,n)
  if num1 == 1 and num2 == 1 and i != 1:
    list1.append(i)
    if len(list1) == 5:
      break
print(list1)

e = int(input('Please choose a public exponent from the above list: '))

start = time.time()

#d = 1
#r = 0

@njit
def modulo(e,t):
    r = 0
    d = 1
    while r != 1:
       r = (d * e) % t
       d = d + 1
    return d - 1

d = modulo(e,t)

# while r != 1:
#   r = (d * e) % t
#   d = d + 1
end = time.time()
print('Runtime: ', (end-start))
print('Modulo: ', n)
print('Public exponent (e): ', e)
print('Private exponent (d): ', d)
```
